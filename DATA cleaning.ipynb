```python
import pandas as pd
import numpy as np
import string
import re




df = pd.read_csv('DATA/USelection2024.csv')
df
```

```python
def clean_text(text):
    # Remove hyperlinks
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Convert text to lowercase
    text = text.lower()

    return text
def remove_emojis(text):
    # Regular expression pattern to match emojis
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & Pictographs
        "\U0001F680-\U0001F6FF"  # Transport & Map Symbols
        "\U0001F1E0-\U0001F1FF"  # Flags
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols & Pictographs
        "\U0001F600-\U0001F64F"  # Additional emoticons
        "\U00002600-\U000026FF"  # Miscellaneous Symbols (e.g., sun, star)
        "\U00002700-\U000027BF"  # Dingbats
        "\U0001F680-\U0001F6FF"  # Transport & Map symbols
        "\U0001F700-\U0001F77F"  # Alchemical symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA00-\U0001FA6F"  # Chess Symbols
        "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        "\U0001FB00-\U0001FBFF"  # Symbols for Legacy Computing
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python
# Clean the text data
df['cleaned_text'] = df['text'].apply(clean_text)

# Remove emojis
df['cleaned_text'] = df['cleaned_text'].apply(remove_emojis)
# code to remove "@" from author names
df['author'] = df['author'].str.replace('@', '', regex=False)

# Export the cleaned text to a new CSV file
df[['author','like_count', 'cleaned_text','hashtags']].to_csv('cleaned_text_data.csv', index=False)

# Display the first few rows to confirm
print(df[['author','like_count', 'cleaned_text','hashtags']].head())
```
